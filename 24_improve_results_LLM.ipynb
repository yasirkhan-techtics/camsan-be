```python
file_path = 'test_drawings/output/output/page32/page_32_light.csv'
df = pd.read_csv(file_path)
print(df.head())
```

```python
GOOGLE_API_KEY = api_key = os.getenv("GOOGLE_API_KEY")
image_path = "test_drawings/output/page_32.png"


# Initialize
resolver = TagOverlapResolver(
    api_key=GOOGLE_API_KEY,
    overlap_threshold=0.9  # 90% overlap
)

# Process
kept_df, removed_df = resolver.process_dataframe(
    df=df,
    image=image_path,
    display_results=True  # Shows each classification
)

```

```python
# Save results
kept_df.to_csv("kept_tags.csv", index=False)
removed_df.to_csv("removed_duplicates.csv", index=False)
```
